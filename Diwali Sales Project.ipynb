**Exploratory Data Analysis**

**Gender**

**Ploting a bar chart for gender and its counts**

```python
ax=sns.countplot(x='Gender',data=a,color='green')
for bars in ax.containers:
  ax.bar_label(bars)
```

**ploting a bar chart for Gender vs total amount**

```python
ss=a.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=ss)
```

**From the above graph we can see that most of the buyers are females and even the purchasing power of females are greter than men**

**Age**

```python
ax=sns.countplot(x='Age Group',data=a,color='salmon',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

** total amount vs age group **

```python
sales_age=a.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

**From the above graph we can see that most of the buyers are from the age group of 26-35 years**

**States**

# total number of orders from the top 10 states

```python
sales_age=a.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders',data=sales_age,color="blue")
```

**from the above graph we can see that most of the orders and total sales\amount are from uttarpardesh,maharastra,karnatka,delhi etc**

**Marital_status**

```python
ac=sns.countplot(data=a,x='Marital_Status')
sns.set(rc={'figure.figsize':(2,5)})
for bars in ac.containers:
  ac.bar_label(bars)
```

```python
sales_age=a.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_age,hue='Gender',color='Blue')
```

**from the above graph we can see that most of the buyers are married(women)and they have high purchasing power**

**Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=a,x='Occupation',color='blue')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Sales_States=a.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=Sales_States)
```

**From the above graph most of the buyers are working in IT sector,Healthcare and aviation center**

**Product category**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=a,x='Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Sales_States=a.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=Sales_States,color='red')
for bars in ax.containers:
  ax.bar_label(bars)

```

**from the above graph  we can see that most of the sold product are clothing and appear electronic categories**

```python
Sales_s=a.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
barplot=sns.barplot(x='Product_ID',y='Orders',data=Sales_s,color='blue')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
fig1,ax1=plt.subplots(figsize=(10,7))
a.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
